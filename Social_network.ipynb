```python
#生成一个自带的网络
g=nx.davis_southern_women_graph()
g.nodes()
```

```python
G=nx.barabasi_albert_graph(1000,2)
pos=nx.spring_layout(G)
nx.draw(G,pos,node_color='red',node_size=60,with_labels=False)
```

```python
node_degree=G.degree()
```

```python
type(node_degree)
```

```python
G.size()
```

```python
node_degree_dict = dict(node_degree)
max_degree_node=max(node_degree_dict,key=node_degree_dict.get)
```

```python
max_degree_node
```

```python
max_degree
```

```python
node_ego=nx.ego_graph(G,max_degree_node)
```

```python
node_ego.degree
```

```python
pos= nx.spring_layout(node_ego)
nx.draw(node_ego, pos, node_color='c', node_size=120,with_labels=False)
```
